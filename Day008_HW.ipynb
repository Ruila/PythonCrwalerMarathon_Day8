### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)
# 想要爬資料的目標網址
r.encoding = 'utf-8'
response = r.text
soup = BeautifulSoup(response, "html5lib")
# print(r.content)
print(type(response))
print(type(soup))
print(soup.prettify())
```
